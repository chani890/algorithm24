### 이 문제는 정의에 의해 반복구조를 이용하여 x를 n번 곱하는 방법으로 간단히 할수 있는데 이것은 억지기법으로, 반복구조로 기술한 x의 n-거듭제곱 알고리즘이다

# 코드 개요
## 입력 변수 
### x : x는 거듭제곱의 밑이다
### n : n는 거듭제곱의 지수이다
## 출력
### 이 알고리즘은 반복문을 이용하여 result에 x를 곱해주는 함수로 x의 n제곱을 출력한다
## 함수설명
### 이 함수는 for문을 통해 x를 n번 곱해주는 함수이다

---
# 코드

```python
def slow_power(x,n):
    result = 1.0
    for i in range(n):
        result = result*x
    return result 
```

---
# 테스트 코드

```python
result = slow_power(5, 2)
print(result)
```
